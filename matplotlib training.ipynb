# OO api in plt

```python
# object oriented method from scratch 
figrue , axes = plt.subplots(figsize=(10,5))

# plot the data
scatter =axes.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target']);

# customise the plots 
axes.set(title='chol v age ',
        xlabel='age',
        ylabel='cholestrol');

# adding a legend

#  the star inside the legend is just doing 'for scatter ,the variable , 
#  unpack all the legend elemnts(),according to the title target ' ...

axes.legend(*scatter.legend_elements(),title='target');


# a horizontal line for this plot or a vertical line , its simple
#            value on wich the line depends and the stat function ,
#        the style of the line ,eg = '-' ,'--','-.',':'  etc
#     the width of the line and its color
axes.axhline((over_50['chol']).mean(),
            linestyle = '--',
            linewidth=2,
            color='navy');
```

```python
# adding a subplot to this so that we get a alook at the chol level and age
# subplotof age , chol , thalach 
figrue , (ax0,ax1)= plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10),
                                sharex = True)

# plotting on axes 0 
scatter = ax0.scatter(x=over_50.age,
            y=over_50.chol,
            c=over_50.target
           );
# customizing plot 1 or ax0
ax0.set(title='chol v age',
       ylabel='chol')

ax0.axhline(over_50.chol.mean(),
           linewidth=2,
           color='red',
           linestyle='-.');

ax0.legend(*scatter.legend_elements(),title='target');

# ax1 --------------------------------------------------------------------------------------------------------------------------
scatter1 = ax1.scatter(x=over_50.age,
            y=over_50.thalach,
            c=over_50.target);

# customizing plot 2
ax1.set(title='thalach v age ',
        xlabel='age',
       ylabel='thalach');

ax1.legend(*scatter1.legend_elements(),title='target');
ax1.set_ylim([70,210]);
ax1.axhline(over_50.thalach.mean(),
           linewidth=3,linestyle='--',
           color='black');


# adding a super title to the entire figure 
figrue.suptitle('Heart Disease Analysis',fontsize=15,fontweight='bold');
```

```python
# subplot of chol age thalach
figure , (ax1, ax2) = plt.subplots(nrows =2, ncols=1,
                                  figsize =(10,10),sharex=True)
scatter1 = ax1.scatter(x=over_50.age
                       ,y=over_50.thalach,
            c=over_50.target);
scatter2 = ax2.scatter(y=over_50.chol,
            x=over_50.age,
           c=over_50.target)

# legend inside the plots
ax1.legend(*scatter.legend_elements(),title='target');
ax2.legend(*scatter1.legend_elements(),title='target');

ax1.set(title='thal v age',ylabel='thalach')
ax2.set(title='chol  v age',ylabel='cholestrol',xlabel='age')

figure.suptitle('HEART DISEASE ANALYSIS',fontsize=30,fontweight='bold');
```
